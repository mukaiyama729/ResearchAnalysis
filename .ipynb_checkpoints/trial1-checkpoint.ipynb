```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
```

```python
df = pd.read_csv('./data/bachelor/data/analysis_result/trial1/data.csv', index_col=0)
df_d = pd.read_csv('./data/bachelor/data/analysis_result/trial1/delaunay.csv', index_col=0)
```

```python
df_d.sort_values('volume', ascending=False).head(100)
Md = df_d.sort_values('volume', ascending=False).head(100)[['x', 'y', 'z']]
```

```python
fig = plt.figure(figsize=(10,10))
M = np.array(df[['x', 'y', 'z']])
x = M[:,0]
y = M[:,1]
z = M[:,2]


xd = Md[:,0]
yd = Md[:,1]
zd = Md[:,2]


ax = fig.add_subplot(projection='3d')
 
#散布図の作成
ax.scatter(x,y,z,s=1,c="blue")
 
#描画
plt.show()
```
