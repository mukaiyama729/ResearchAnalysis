```python
import pathlib
import shutil
```

```python
def change_suffix(file_name, from_suffix, to_suffix, path_after_change):
    # ファイルの拡張子を得る
    sf = pathlib.PurePath(file_name).suffix
    
    # 変更対象かどうか判定する
    if sf == from_suffix:
        # ファイル名(拡張子除く)を得る
        st = pathlib.PurePath(file_name).stem

        # 変更後のファイル名を得る
        to_name = st + to_suffix

        # ファイル名を変更する
        shutil.move(file_name, path_after_change + to_name)
```

```python
import glob
import pathlib
import shutil

def change_suffix(file_name, from_suffix, to_suffix, path_after_change):
    # ファイルの拡張子を得る
    sf = pathlib.PurePath(file_name).suffix
    
    # 変更対象かどうか判定する
    if sf == from_suffix:
        # ファイル名(拡張子除く)を得る
        st = pathlib.PurePath(file_name).stem

        # 変更後のファイル名を得る
        to_name = st + to_suffix

        # ファイル名を変更する
        shutil.move(file_name, path_after_change + to_name)

def list_of_files(directory_path, suffix):
    files = glob.glob(os.path.join(directory_path, '*'+ suffix))
    return files

def change_all_suffix(directory_path, path_after_change, from_suffix, to_suffix):
    for path in list_of_files(directory_path, from_suffix):
        change_suffix(path, from_suffix, to_suffix, path_after_change)
```

```python
directory_path = '/Users/mukaiyama/research/kitao/Delaunay/data/bachelor/data/MDM2p53TAD3DComCoord/'
path_after_change = '/Users/mukaiyama/research/kitao/Delaunay/data/bachelor/data/txt_files/'
from_suffix = '.xvg'
to_suffix = '.txt'
change_all_suffix(directory_path, path_after_change, from_suffix, to_suffix)
```

```python
def change_suffix(file_name, from_suffix, to_suffix, path_after_change):
    # ファイルの拡張子を得る
    sf = pathlib.PurePath(file_name).suffix
    
    # 変更対象かどうか判定する
    if sf == from_suffix:
        # ファイル名(拡張子除く)を得る
        st = pathlib.PurePath(file_name).stem

        # 変更後のファイル名を得る
        to_name = st + to_suffix

        # ファイル名を変更する
        shutil.copy(file_name, path_after_change + to_name)
```
